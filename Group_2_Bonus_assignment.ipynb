```python
#Afua Frempong 90434176
#Shreya Saxena 41969981

import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

url = 'https://www.planecrashinfo.com/1920/1920.htm'
response = requests.get(url)


if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.find('table')

    # Extract data from the table
    data = []
    for row in table.find_all('tr'):
        row_data = [cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])]
        data.append(row_data)

    # Create a DataFrame
    df = pd.DataFrame(data[1:], columns=data[0])
    df.to_csv('plane_crash_data.csv', index=False)

    #Perform your analysis with at least 2 libraries that were used in class. Details must be provided as to what you used.
    #You need to present/describe 4 things about the data. The results will need to look at relationships in the data (ie. a map of the top 5 worst air qualities in BC this last year (mean of max air vs lat/long).
    #Results can be presented in tabluar and graphic form (chart, map, etc)

    df['Fatalities'] = df['Fatalities'].str.extract('(\d+)/').astype(float)

    import matplotlib.pyplot as plt

    # Group by 'Location/Operator' and sum the fatalities. This will tell us which operator has the most fatalities
    grouped_data = df.groupby('Location / Operator')['Fatalities'].sum().sort_values(ascending=False)

    # Plot
    grouped_data.head(10).plot(kind='bar', figsize=(10, 6))  # Top 10 for clarity
    plt.title('Top 10 Locations/Operators vs. Number of Deaths')
    plt.xlabel('Location / Operator')
    plt.ylabel('Number of Deaths')
    plt.xticks(rotation=45)
    plt.show()

    # Count the frequency of each 'Aircraft Type / Registration' per 'Location / Operator'
    grouped_data_type = df.groupby(['Location / Operator', 'Aircraft Type / Registration']).size().reset_index(name='Counts')

    # Selecting top 5 locations/operators based on the number of incidents. What operators have the most incidents
    top_locations = df['Location / Operator'].value_counts().head(5).index
    filtered_data = grouped_data_type[grouped_data_type['Location / Operator'].isin(top_locations)]

    # Plot
    for location in top_locations:
        subset = filtered_data[filtered_data['Location / Operator'] == location]
        subset = subset.sort_values(by='Counts', ascending=False).head(5)  # Top 5 aircraft types for clarity
        subset.plot(x='Aircraft Type / Registration', y='Counts', kind='bar', title=location)
        plt.ylabel('Counts')
        plt.xticks(rotation=45)
        plt.show()

#Aircraft type and number fatalities per type
# Group the data by aircraft type and sum the casualties. We want to know which aircrafts have the most incidents
    casualties_by_aircraft = df.groupby('Aircraft Type / Registration')['Fatalities'].sum().reset_index()

# Sort the data by total fatalities in descending order
    casualties_by_aircraft = casualties_by_aircraft.sort_values(by='Fatalities', ascending=False)

    plt.figure(figsize=(12, len(casualties_by_aircraft) * 0.5))
    plt.barh(casualties_by_aircraft['Aircraft Type / Registration'], casualties_by_aircraft['Fatalities'], color='skyblue')
    plt.xlabel('Total Fatalities')
    plt.ylabel('Aircraft Type')
    plt.title('Total Fatalities by Aircraft Type')
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    #Fatality percentage in every given month for the year 1920
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

    df['Month'] = df['Date'].dt.month_name()
    fatalities_by_month = df.groupby('Month')['Fatalities'].sum().reset_index()

    # Sort the data by month order
    month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    fatalities_by_month['Month'] = pd.Categorical(fatalities_by_month['Month'], categories=month_order, ordered=True)
    fatalities_by_month = fatalities_by_month.sort_values('Month')

    plt.figure(figsize=(8, 8))
    plt.pie(fatalities_by_month['Fatalities'], labels=fatalities_by_month['Month'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title('Total Fatalities by Month')
    fig = px.pie(fatalities_by_month, names='Month', values='Fatalities', title='Total Fatalities by Month',
                 labels={'Total Fatalities': 'Fatalities'},
                 color_discrete_sequence=px.colors.qualitative.Set3,
                 template='plotly_dark')

    fig.show()

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import matplotlib.pyplot as plt
import plotly.express as px


def clean_and_convert_fatalities(value):
    match = re.search(r'\d+', value)
    return int(match.group()) if match else 0


def scrape_and_save_data(url, csv_filename):
    response = requests.get(url)


    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table')

        data = []
        for row in table.find_all('tr'):
            row_data = [cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])]
            data.append(row_data)

        df = pd.DataFrame(data[1:], columns=data[0])
        df.to_csv(csv_filename, index=False)

# URLs for 1920 and 2020
url_1920 = 'https://www.planecrashinfo.com/1920/1920.htm'
url_2020 = 'https://www.planecrashinfo.com/2020/2020.htm'


csv_filename_1920 = 'plane_crash_data1920.csv'
csv_filename_2020 = 'plane_crash_data2020.csv'

# Scrape and save data for 1920 and 2020
scrape_and_save_data(url_1920, csv_filename_1920)
scrape_and_save_data(url_2020, csv_filename_2020)

df_1920 = pd.read_csv(csv_filename_1920)
df_2020 = pd.read_csv(csv_filename_2020)

df_1920['Total Fatalities'] = df_1920['Fatalities'].apply(clean_and_convert_fatalities)
df_2020['Total Fatalities'] = df_2020['Fatalities'].apply(clean_and_convert_fatalities)


fatalities_1920 = df_1920['Total Fatalities'].sum()
fatalities_2020 = df_2020['Total Fatalities'].sum()


comparison_df = pd.DataFrame({'Year': ['1920', '2020'],
                              'Total Fatalities': [fatalities_1920, fatalities_2020]})

fig = px.bar(comparison_df, x='Year', y='Total Fatalities', title='Total Fatalities in 1920 vs. 2020',
             labels={'Total Fatalities': 'Fatalities'}, color='Year',
             template='plotly_dark')

fig.show()

```
